```python
# File: src/calibrator.py

import pandas as pd
import numpy as np

def calibrate_model1_by_dt(events_df: pd.DataFrame):
    """
    Model I 校准（纯 dt 方法）：
      Λ̂_i(n) = 1 / mean(dt | side,level,depth=n)
      λ̂^e_i(n) = Λ̂_i(n) * N_e(n) / N(n)
    返回 DataFrame 列：
      side, level, depth,
      Lambda, lambda_L, lambda_C, lambda_M,
      N_total, N_L, N_C, N_M
    """
    df = events_df[events_df['event'].isin(['L','C','M'])].copy()
    df = df.sort_values('time')
    # 计算同 side,level 的 dt
    df['dt'] = df.groupby(['side','level'])['time'].diff()
    # 平均 dt
    dt_mean = (df.dropna(subset=['dt'])
                 .groupby(['side','level','depth'])['dt']
                 .mean()
                 .rename('dt_mean'))
    Lambda = (1.0 / dt_mean).rename('Lambda')
    # 统计 N_L, N_C, N_M
    cnt = (df.groupby(['side','level','depth','event'])
             .size()
             .unstack(fill_value=0)
             .rename(columns={'L':'N_L','C':'N_C','M':'N_M'}))
    cnt['N_total'] = cnt[['N_L','N_C','N_M']].sum(axis=1)
    # 合并
    rates = pd.concat([Lambda, cnt], axis=1).fillna(0)
    rates['lambda_L'] = rates['Lambda'] * rates['N_L']/rates['N_total'].replace(0,1)
    rates['lambda_C'] = rates['Lambda'] * rates['N_C']/rates['N_total'].replace(0,1)
    rates['lambda_M'] = rates['Lambda'] * rates['N_M']/rates['N_total'].replace(0,1)
    return rates.reset_index()[[
        'side','level','depth',
        'Lambda','lambda_L','lambda_C','lambda_M',
        'N_total','N_L','N_C','N_M'
    ]]

def add_context(events_df: pd.DataFrame, quantiles=(0.33,0.66)):
    """
    为每条事件添加对侧 L1 深度 context：
      context ∈ {'Q0','Q-','Q¯','Q+'}
    返回 (带 context 列的 df, (m,l) 阈值)
    """
    depths = events_df['opp_depth1']
    m = depths.quantile(quantiles[0])
    l = depths.quantile(quantiles[1])
    def ctx(x):
        if x == 0:      return 'Q0'
        elif x <= m:    return 'Q-'
        elif x <= l:    return 'Q¯'
        else:           return 'Q+'
    df = events_df.copy()
    df['context'] = df['opp_depth1'].apply(ctx)
    return df, (m, l)

def calibrate_intensity_exact_ctx(events_df: pd.DataFrame):
    """
    Model II 带 context 校准（dt 方法）：
      Λ̂, λ̂^e per (side,level,context,depth)
    返回 DataFrame 列：
      side, level, context, depth,
      Lambda, lambda_L, lambda_C, lambda_M,
      N_total, N_L, N_C, N_M
    """
    df = events_df[events_df['event'].isin(['L','C','M'])].copy()
    df = df.sort_values('time')
    df['dt'] = df.groupby(['side','level','context'])['time'].diff()
    dt_mean = (df.dropna(subset=['dt'])
                 .groupby(['side','level','context','depth'])['dt']
                 .mean()
                 .rename('dt_mean'))
    Lambda = (1.0 / dt_mean).rename('Lambda')
    cnt = (df.groupby(['side','level','context','depth','event'])
             .size()
             .unstack(fill_value=0)
             .rename(columns={'L':'N_L','C':'N_C','M':'N_M'}))
    cnt['N_total'] = cnt[['N_L','N_C','N_M']].sum(axis=1)
    rates = pd.concat([Lambda, cnt], axis=1).fillna(0)
    rates['lambda_L'] = rates['Lambda'] * rates['N_L']/rates['N_total'].replace(0,1)
    rates['lambda_C'] = rates['Lambda'] * rates['N_C']/rates['N_total'].replace(0,1)
    rates['lambda_M'] = rates['Lambda'] * rates['N_M']/rates['N_total'].replace(0,1)
    cols = ['side','level','context','depth',
            'Lambda','lambda_L','lambda_C','lambda_M',
            'N_total','N_L','N_C','N_M']
    return rates.reset_index()[cols]
```

```python
# File: src/qbd_solver.py

import numpy as np
import pandas as pd
from matrixgeometric import qbd_stationary

def build_q_blocks(distr_ctx: pd.DataFrame, ctx_col='context'):
    """
    构建 Q0,Q1,Q2 转移矩阵块，用于 QBD solver。
    """
    max_depth = int(distr_ctx['depth'].max())
    ctx_vals  = sorted(distr_ctx[ctx_col].unique())
    M = len(ctx_vals)
    S = (max_depth+1)*M
    Q0 = np.zeros((S,S)); Q1 = np.zeros((S,S)); Q2 = np.zeros((S,S))
    def idx(n,c): return n*M + ctx_vals.index(c)
    for _, row in distr_ctx.iterrows():
        n    = int(row['depth'])
        c    = row[ctx_col]
        i    = idx(n,c)
        lamL = float(row['lambda_L'])
        lamD = float(row['lambda_C'] + row['lambda_M'])
        Q0[i,i] -= (lamL + lamD)
        if n < max_depth:
            Q1[i, idx(n+1,c)] = lamL
        else:
            Q0[i,i] += lamL
        if n > 0:
            Q2[i, idx(n-1,c)] = lamD
        else:
            Q0[i,i] += lamD
    return Q0, Q1, Q2, ctx_vals, max_depth

def solve_model2_qbd(distr_ctx: pd.DataFrame,
                     side: str,
                     level: int,
                     ctx_col: str = 'context'):
    """
    求 Model II QBD 平稳分布 π(c,n)（单个 side,level）。
    返回 DataFrame 列:
      side, level, context, depth, pi
    """
    sub = distr_ctx[(distr_ctx['side']==side) &
                    (distr_ctx['level']==level)].copy()
    if sub.empty:
        return pd.DataFrame(columns=['side','level',ctx_col,'depth','pi'])
    Q0,Q1,Q2,ctx_vals,max_depth = build_q_blocks(sub, ctx_col)
    pi_vec = qbd_stationary(Q0,Q1,Q2)
    recs = []
    M = len(ctx_vals)
    for n in range(max_depth+1):
        for j,c in enumerate(ctx_vals):
            recs.append({
                'side': side,
                'level': level,
                ctx_col: c,
                'depth': n,
                'pi':   float(pi_vec[n*M + j])
            })
    return pd.DataFrame(recs)
```

```python
# File: src/minute_mid.py

import pandas as pd
from collections import defaultdict

def build_minute_midprice(message_file: str) -> pd.Series:
    """
    从 message-only 重放簿构建每分钟 mid-price 序列。
    返回 pd.Series，index=minute, value=mid.
    """
    cols = ['time','type','order_id','size','price','direction']
    df = pd.read_csv(message_file, header=None, names=cols)
    df = df[df['type'].isin([1,2,3,4])].sort_values('time')

    bid_book = defaultdict(int)
    ask_book = defaultdict(int)
    mids = []
    for _, r in df.iterrows():
        tp   = int(r['type'])
        ev   = 'L' if tp==1 else ('C' if tp in (2,3) else 'M')
        side = 'bid' if r['direction']==1 else 'ask'
        book = bid_book if side=='bid' else ask_book
        p, sz = r['price'], int(r['size'])
        if ev=='L':
            book[p] += sz
        else:
            book[p] = max(book[p]-sz, 0)
        if bid_book and ask_book:
            mid = (max(bid_book.keys()) + min(ask_book.keys()))/2.0
            mids.append((r['time'], mid))

    mid_df   = pd.DataFrame(mids, columns=['time','mid']).set_index('time')
    mid_min  = mid_df['mid'].resample('60S', origin='start').last().dropna()
    mid_min.index = range(len(mid_min))
    mid_min.index.name = 'minute'
    return mid_min
```

```python
# File: src/theta_fit_exact.py

import numpy as np
import pandas as pd
from scipy.optimize import least_squares
from simulator_model3 import SimConfig, simulate_model3_paths

def realised_vol_minute(mid: pd.Series) -> float:
    r = np.log(mid/mid.shift(1)).dropna()
    return float(r.std())

def mean_reversion_eta(mid: pd.Series, dt_minutes: int = 1) -> float:
    dp = mid.diff().dropna()
    rho = dp.autocorr(lag=1)
    return 0.0 if rho is None or rho<=0 else -np.log(rho)/dt_minutes

def fit_theta_theta_reinit_exact(mid_hist: pd.Series,
                                 lambdas: pd.DataFrame,
                                 pi_qbd: pd.DataFrame,
                                 tick: float,
                                 m_thresh: float,
                                 l_thresh: float,
                                 minutes:int=240,
                                 paths:int=500,
                                 seed:int=0):
    """
    通过最小二乘拟合 θ, θ_reinit，使模拟 mid 的 σ, η 与实证匹配。
    """
    sigma_emp = realised_vol_minute(mid_hist)
    eta_emp   = mean_reversion_eta(mid_hist)
    def obj(p):
        th, tr = p
        if not (0<th<1 and 0<=tr<1):
            return np.array([1e3,1e3])
        cfg = SimConfig(lambdas, pi_qbd, tick, th, tr,
                        minutes, paths, seed,
                        m_thresh, l_thresh)
        mids = simulate_model3_paths(cfg, return_full_book=False)
        sig = [realised_vol_minute(mids[col]) for col in mids.columns]
        et  = [mean_reversion_eta(mids[col])    for col in mids.columns]
        return np.array([np.mean(sig)-sigma_emp,
                         np.mean(et)-eta_emp])
    res = least_squares(obj, x0=[0.1,0.1],
                        bounds=([1e-4,0],[0.999,0.999]))
    theta, theta_reinit = res.x
    return theta, theta_reinit, pd.DataFrame([{
        'theta':theta,'theta_reinit':theta_reinit,
        'res_norm':np.linalg.norm(res.fun),
        'success':res.success
    }])
```

```python
# File: src/run_full.py

import os
import pandas as pd

from extractor      import extract_events_msg_ob
from calibrator     import calibrate_model1_by_dt, add_context, calibrate_intensity_exact_ctx
from qbd_solver     import solve_model2_qbd
from minute_mid     import build_minute_midprice
from theta_fit_exact import fit_theta_theta_reinit_exact
# Model III imports omitted

# ------------- CONFIG -------------
MESSAGE_FILE = 'data/LOBSTER/ABC_message.csv'
ORDERBOOK_FILE= 'data/LOBSTER/ABC_orderbook.csv'
K_LEVELS     = 5
OUT_DIR      = 'data/processed_events'
TICK         = 0.01
# θ 与 simulate 参数
MINUTES_THETA=240
PATHS_THETA  =500
# ----------------------------------

os.makedirs(OUT_DIR, exist_ok=True)

# 1) Extract events
ev, tick = extract_events_msg_ob(MESSAGE_FILE, ORDERBOOK_FILE, K=K_LEVELS)
ev.to_csv(f'{OUT_DIR}/events.csv', index=False)

# 2) Model I calibration
rates_m1 = calibrate_model1_by_dt(ev)
rates_m1.to_csv(f'{OUT_DIR}/rates_model1.csv', index=False)

# 3) Add context & Model II calibration
ev_ctx, (m_thresh,l_thresh) = add_context(ev)
rates_m2 = calibrate_intensity_exact_ctx(ev_ctx)
rates_m2.to_csv(f'{OUT_DIR}/rates_model2.csv', index=False)

# 4) QBD stationary for Model II
pi2 = solve_model2_qbd(rates_m2, side='bid', level=1)  # example for bid, lvl1
pi2.to_csv(f'{OUT_DIR}/pi_model2_bid1.csv', index=False)

# 5) Empirical minute mid
mid = build_minute_midprice(MESSAGE_FILE)
mid.to_csv(f'{OUT_DIR}/minute_mid.csv')

# 6) Fit θ & θ_reinit
theta, theta_reinit, diag = fit_theta_theta_reinit_exact(
    mid, rates_m2, pi2, TICK, m_thresh, l_thresh,
    minutes=MINUTES_THETA, paths=PATHS_THETA, seed=0
)
diag.to_csv(f'{OUT_DIR}/theta_fit.csv', index=False)
print('theta, theta_reinit =', theta, theta_reinit)

print("=== Pipeline complete ===")
```
