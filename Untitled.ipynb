```python
import pandas as pd
import numpy as np
from collections import defaultdict
import math

def infer_tick_size_full(prices: np.ndarray) -> int:
    """
    精确用全量价格差的最大公约数推断 tick_size。
    """
    arr = np.sort(np.unique(prices))
    diffs = np.diff(arr)
    diffs = diffs[diffs > 0]
    return int(np.gcd.reduce(diffs.astype(int))) if len(diffs) else 1

def compute_pmid(bid1: float,
                 ask1: float,
                 tick: float,
                 prev_pmid: float = None) -> float:
    """
    按 HL&R parity 规则计算参考价 θ(t):
      - 奇数 spread → θ = mid
      - 偶数 spread → θ 向 prev_pmid 方向偏移 ±0.5*tick
    """
    mid = (ask1 + bid1) / 2.0
    spread_ticks = int(round((ask1 - bid1) / tick))
    if spread_ticks % 2 == 1 or prev_pmid is None:
        return mid
    return mid - 0.5*tick if prev_pmid <= mid else mid + 0.5*tick

def price_to_level_theta(side: str,
                         price: int,
                         theta: float,
                         tick: float) -> int:
    """
    用 θ + k*δ 定义档位：
      bid 侧: level = floor((θ - price)/δ) + 1
      ask 侧: level = floor((price - θ)/δ) + 1
    返回 <1 或 >K 的需丢弃。
    """
    if side == 'bid':
        d = (theta - price) / tick
    else:
        d = (price - theta) / tick
    if d < 0:
        return -1
    return math.ceil(d)

def extract_events(message_file: str,
                   snapshot_file: str,
                   K: int = 5,
                    max_level: int =3):
    """
    行对行对齐的消息+快照Extractor。

    Parameters
    ----------
    message_file : str
        CSV路径，列 ['time','type','order_id','size','price','direction']
    snapshot_file : str
        CSV路径，每行为消息执行前的前K档orderbook
    K : int
        前K档

    Returns
    -------
    events_df : pandas.DataFrame
        列 ['time','side','level','depth','opp_depth1','event','dt','pmid']
    tick : float
        推断出的 tick size
    """
    # 1) 读取文件
    msg = pd.read_csv(message_file, header=None,
                      names=['time','type','order_id','size','price','direction'])
    ob  = pd.read_csv(snapshot_file, header=None)
    # 2) 给 snapshot 列命名
    cols = []
    for L in range(1, K+1):
        cols += [f'ask_price_{L}', f'ask_size_{L}',
                 f'bid_price_{L}', f'bid_size_{L}']
    ob.columns = cols[:ob.shape[1]] + ob.columns.tolist()[len(cols):]

    # 3) 推断 tick size
    prices_msg = msg['price'].to_numpy()
    price_cols = [c for c in ob.columns if 'price' in c]
    prices_ob  = ob[price_cols].to_numpy().ravel()
    tick = infer_tick_size_full(np.concatenate([prices_msg, prices_ob]))

    events = []
    last_t = None
    prev_pmid = None

    # 4) 遍历每行
    for i in range(len(msg)):
        m = msg.iloc[i]
        tp = int(m['type'])
        if tp not in (1,2,3,4):
            continue

        # 4.0 取“执行前状态”快照行
        if i == 0:
            pre_snap = ob.iloc[0]        # 没有 i−1 时，用第0行作为初始
        else:
            pre_snap = ob.iloc[i-1]

        # 4.1 读消息字段
        t     = float(m['time'])
        price = float(m['price'])
        size  = int(m['size'])
        side  = 'bid' if m['direction']==1 else 'ask'
        ev    = 'L' if tp==1 else ('C' if tp in (2,3) else 'M')

        # 4.2 构建快照中的簿状态字典（含 size=0）
        bid_book = {
            float(pre_snap[f'bid_price_{L}']): int(pre_snap[f'bid_size_{L}'])
            for L in range(1, K+1)
        }
        ask_book = {
            float(pre_snap[f'ask_price_{L}']): int(pre_snap[f'ask_size_{L}'])
            for L in range(1, K+1)
        }

        # 4.3 计算 pmid（θ）
        bid1 = float(pre_snap['bid_price_1'])
        ask1 = float(pre_snap['ask_price_1'])
        pmid = compute_pmid(bid1, ask1, tick, prev_pmid)
        prev_pmid = pmid

        # 4.4 depth_before
        depth_before = (bid_book if side=='bid' else ask_book).get(price, 0)

        # 4.5 level via θ-grid
        level = price_to_level_theta(side, price, pmid, tick)

        # 4.6 opp_depth1
        opp_book = ask_book if side=='bid' else bid_book
        if side == 'bid':
            opp_price1 = min(opp_book.keys())
        else:
            opp_price1 = max(opp_book.keys())
        opp_depth1 = opp_book.get(opp_price1, 0)

        # 4.7 记录仅当 1≤level≤max_level
        if 1 <= level <= max_level:
            dt = t - last_t if last_t is not None else 0.0
            events.append({
                'time':       t,
                'side':       side,
                'level':      level,
                'depth':      depth_before,
                'opp_depth1': opp_depth1,
                'event':      ev,
                'dt':         dt,
                'pmid':       pmid
            })
            last_t = t

    return pd.DataFrame(events), tick
```

```python
import pandas as pd

def calibrate_exact_intensities(events_df: pd.DataFrame):
    
    # 只保留我们关心的三类事件
    df = events_df[events_df['event'].isin(['L','C','M'])].copy()

    # 1) 计算每个 (side,level,depth) 的平均 dt → Λ̂ = 1/mean(dt)
    dt_mean = df.groupby(['side','level','depth'])['dt'].mean().rename('dt_mean')
    Lambda = (1.0 / dt_mean).rename('Lambda')

    # 2) 统计各类事件数
    cnt = (
        df
        .groupby(['side','level','depth','event'])
        .size()
        .unstack(fill_value=0)
        .rename(columns={'L':'N_L','C':'N_C','M':'N_M'})
    )
    cnt['N_total'] = cnt[['N_L','N_C','N_M']].sum(axis=1)

    # 3) 合并并计算分类型强度
    rates = pd.concat([Lambda, cnt], axis=1).fillna(0)
    rates['lambda_L'] = rates['Lambda'] * rates['N_L'] / rates['N_total'].replace(0,1)
    rates['lambda_C'] = rates['Lambda'] * rates['N_C'] / rates['N_total'].replace(0,1)
    rates['lambda_M'] = rates['Lambda'] * rates['N_M'] / rates['N_total'].replace(0,1)

    # 4) 整理输出顺序
    rates_df = (
        rates
        .reset_index()[[
            'side','level','depth',
            'Lambda','lambda_L','lambda_C','lambda_M',
            'N_total','N_L','N_C','N_M'
        ]]
    )
    return rates_df
```

```python
import pandas as pd
import numpy as np

def solve_stationary_from_rates(rates_df: pd.DataFrame, side: str, level: int):
    """
    依据 calibrate_exact_intensities 输出 rates_df 计算稳态分布 π_i(n)
    for a given side and level.
    
    parameters
    ----------
    rates_df : DataFrame
        必须包含列 ['side','level','depth','lambda_L','lambda_C','lambda_M']
    side : str
        'bid' 或 'ask'
    level : int
        队列层级 i
    
    returns
    -------
    pi_df : DataFrame
        包含 ['depth', 'pi']，depth 从 0 到 N_max，pi 为稳态概率
    """
    # 1) 选取该 side 和 level 的子集，并按 depth 排序
    sub = rates_df[(rates_df['side'] == side) & (rates_df['level'] == level)].copy()
    sub = sub.sort_values('depth').reset_index(drop=True)
    n_values = sub['depth'].values.astype(int)
    
    # 2) 最大深度
    N_max = n_values.max()
    
    # 3) 建立字典以便快速查找
    lam_L_map = dict(zip(sub['depth'], sub['lambda_L']))
    lam_C_map = dict(zip(sub['depth'], sub['lambda_C']))
    lam_M_map = dict(zip(sub['depth'], sub['lambda_M']))
    
    # 4) 递推计算 rho(n) 和 π(n)
    pi = np.zeros(N_max + 1)
    pi[0] = 1.0  # 未规范化的 π(0)
    for n in range(1, N_max + 1):
        lam = lam_L_map.get(n - 1, 0.0)
        mu  = lam_C_map.get(n, 0.0) + lam_M_map.get(n, 0.0)
        rho = lam / mu if mu > 0 else 0.0
        pi[n] = pi[n - 1] * rho
    
    # 5) 规范化
    total = pi.sum()
    if total > 0:
        pi /= total
    
    # 6) 返回 DataFrame
    pi_df = pd.DataFrame({
        'depth': np.arange(N_max + 1, dtype=int),
        'pi':    pi
    })
    return pi_df

# Example usage:
# rates_df = calibrate_exact_intensities(events_df)
# pi_bid1 = solve_stationary_from_rates(rates_df, side='bid', level=1)
# print(pi_bid1.head())
# plt.plot(pi_bid1['depth'], pi_bid1['pi'], marker='o'); plt.show()
```

```python
def solve_model1_stationary(lambda_L, lambda_C, lambda_M, side: str, lvl: int, N_max: int = 50):
    """
    Compute stationary distribution π_i(n) for Model I queue (side, lvl),
    up to depth n=0..N_max, with birth = λ_L, death = λ_C+λ_M.

    Returns:
      pi : list of length N_max+1 with π(0..N_max), sum ≈1.
    """
    # compute ρ(n) = λ_L(n-1) / μ(n), where μ(n)=λ_C(n)+λ_M(n)
    rho = [0.0] * (N_max+1)
    for n in range(1, N_max+1):
        lam = lambda_L.get((side,lvl,n-1), 0.0)
        mu  = lambda_C.get((side,lvl,n),0.0) + lambda_M.get((side,lvl,n),0.0)
        rho[n] = lam/mu if mu>0 else 0.0

    # π(0)
    pi = [1.0]
    # compute π(n)=π(0)*∏_{k=1}^n ρ(k)
    for n in range(1, N_max+1):
        pi.append(pi[-1] * rho[n])

    # normalize
    S = sum(pi)
    if S>0:
        pi = [p/S for p in pi]
    return pi
```

```python
def add_context(events_df: pd.DataFrame):
    pos = events_df.loc[events_df['opp_depth1']>0, 'opp_depth1']
    m = float(pos.quantile(0.33)) if len(pos)>0 else 0.0
    l = float(pos.quantile(0.66)) if len(pos)>0 else 0.0
    def bucket(x):
        if x==0: return 'Q0'
        elif x<=m: return 'Q-'
        elif x<=l: return 'Qbar'
        else: return 'Q+'
    df = events_df.copy()
    df['context'] = df['opp_depth1'].apply(bucket)
    return df, (m,l)
```

```python
### FILE: src/qbd_solver.py
import numpy as np
import pandas as pd
from matrixgeometric import qbd_stationary

def build_q_blocks(distr_ctx: pd.DataFrame, ctx_col='context', max_depth=None):
    if max_depth is None:
        max_depth = int(distr_ctx['depth'].max())
    ctx_vals = sorted(distr_ctx[ctx_col].unique())
    M = len(ctx_vals); S = (max_depth+1)*M
    Q0 = np.zeros((S,S)); Q1 = np.zeros((S,S)); Q2 = np.zeros((S,S))
    def idx(n,m): return n*M + ctx_vals.index(m)
    for n in range(max_depth+1):
        for m in ctx_vals:
            r = distr_ctx[(distr_ctx['depth']==n) & (distr_ctx[ctx_col]==m)]
            lamL = float(r['lambda_L'].iloc[0]) if not r.empty else 0.0
            lamD = float((r['lambda_C']+r['lambda_M']).iloc[0]) if not r.empty else 0.0
            i = idx(n,m)
            Q0[i,i] -= (lamL + lamD)
            if n < max_depth: Q1[i, idx(n+1,m)] += lamL
            else:             Q0[i,i] += lamL
            if n > 0:         Q2[i, idx(n-1,m)] += lamD
            else:             Q0[i,i] += lamD
    return Q0,Q1,Q2,ctx_vals,max_depth

def solve_model2_qbd(distr_ctx: pd.DataFrame, side: str, level: int, ctx_col='context') -> pd.DataFrame:
    sub = distr_ctx[(distr_ctx['side']==side) & (distr_ctx['level']==level)]
    if sub.empty:
        return pd.DataFrame(columns=['side','level','depth',ctx_col,'pi'])
    Q0,Q1,Q2,ctx_vals,max_depth = build_q_blocks(sub, ctx_col)
    pi_vec = qbd_stationary(Q0,Q1,Q2)
    rows=[]; M=len(ctx_vals)
    for n in range(max_depth+1):
        for mi,m in enumerate(ctx_vals):
            rows.append({'side':side,'level':level,'depth':n,ctx_col:m,'pi': float(pi_vec[n*M+mi])})
    return pd.DataFrame(rows)

def solve_qbd_all(distr_ctx: pd.DataFrame, ctx_col='context') -> pd.DataFrame:
    res=[]
    for (s,lvl), _ in distr_ctx.groupby(['side','level']):
        res.append(solve_model2_qbd(distr_ctx, s, lvl, ctx_col))
    return pd.concat(res, ignore_index=True)
```
